## Question 1

```python
import numpy as np

def is_linear_function(func):
    # Check additivity
    additivity_check = True
    for x1, x2, y1, y2 in [(1, 2, 4, 7), (0.5, 0.8, 2, 3), (-3, 1, 5, -2)]:
        additivity_check = additivity_check and np.all(func(x1 + x2, y1 + y2) == tuple(val1 + val2 for val1, val2 in zip(func(x1, y1), func(x2, y2))))

    # Check homogeneity
    homogeneity_check = True
    for k, x, y in [(2, 1, 3), (0.5, -2, 4), (-1, 0, 5)]:
        homogeneity_check = homogeneity_check and np.all(func(k * x, k * y) == tuple(k * val for val in func(x, y)))

    return additivity_check and homogeneity_check

# Define the transformations
function_a = lambda x, y: (2 * x + y, 3 * x + 4 * y)
function_b = lambda x, y: (np.sin(x + y), np.sin(x - y))
function_c = lambda x, y: (x * y, y)
function_d = lambda x, y: (3 * x - y, y, x)

# Check if each transformation is linear
is_linear_a = is_linear_function(function_a)
is_linear_b = is_linear_function(function_b)
is_linear_c = is_linear_function(function_c)
is_linear_d = is_linear_function(function_d)

# Print the results
print("Is transformation (a) linear?", is_linear_a)
print("Is transformation (b) linear?", is_linear_b)
print("Is transformation (c) linear?", is_linear_c)
print("Is transformation (d) linear?", is_linear_d)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Original cat data
topx = np.array([1, .98, .8, .83, .8, .77, .5, .25, 0, -.25, -.5, -.8,
                  -.8, -.98, -1])
topy = np.array([0, .15, .7, 1.15, 1.03, 1.15, .85, .95, 1, .95, .85,
                  1.2, .7, .15, 0])
botx = -np.cos(np.pi * np.arange(1, 10) / 10)
boty = -np.sin(np.pi * np.arange(1, 10) / 10)
wiskx = np.array([.2, 1.3, .2, 1.4, .2, 1.4, .2, 1.3, .2, .17, .13, .08, .03, 0])
wisky = np.array([0, .3, 0, .1, 0, -.1, 0, -.3, 0, .1, -.1, .1, -.1, 0]) - 0.2
xeye = np.array([0, .2, .3, .4, .43, .45, .43, .4, .37, .35, .37, .4, .5,
                 .6, .5, .4, .3, .2, 0])
yeye = np.array([.5, .5, .43, .4, .42, .5, .58, .6, .58, .5, .42, .4, .43,
                 .5, .57, .6, .57, .5, .5]) - 0.2
x = np.concatenate((topx, botx, wiskx, xeye, -xeye, -wiskx[::-1]))
y = np.concatenate((topy, boty, wisky, yeye, yeye, wisky[::-1]))
cat = np.array([x, y]).T

# Define the reflection through the line function
def reflection_through_line(u, data):
    n = len(u)
    # Calculate the transformation matrix using the formula 2uu^T - I
    identity_matrix = np.identity(n)
    transformation_matrix = 2 * np.outer(u, u) - identity_matrix

    # Apply the transformation to the data
    transformed_data = np.dot(data, transformation_matrix)

    return transformed_data

# Create a 4x4 grid of subplots
fig, axs = plt.subplots(5, 4, figsize=(12, 12))

# Apply transformations and plot them in the subplots
transformations = [
    ("Original", cat),

    ("Reflection through the x-axis", reflection_through_line(np.array([1, 0]), cat)),

    ("Reflection through the y-axis", reflection_through_line(np.array([0, 1]), cat)),

    ("Reflection through the axis y=x", reflection_through_line(np.array([1 / np.sqrt(2), 1 / np.sqrt(2)]), cat)),

    ("Reflection through the axis y=-x", reflection_through_line(np.array([1 / np.sqrt(2), -1 / np.sqrt(2)]), cat)),

    ("Reflection through the origin", reflection_through_line(np.array([0, 0]), cat)),

    ("Horizontal Contraction by a factor 1/3", np.dot(cat, np.array([[1/3, 0], [0, 1]]))),

    ("Vertical Contraction by a factor 1/5", np.dot(cat, np.array([[1, 0], [0, 1/5]]))),

    ("Horizontal Expansions by a factor 3", np.dot(cat, np.array([[3, 0], [0, 1]]))),

    ("Vertical Expansions by a factor of 4", np.dot(cat, np.array([[1, 0], [0, 4]]))),

     ("Horizontal Shear by a factor of -1.5", np.dot(cat, np.array([[1, -1.5], [0, 1]]))),

    ("Horizontal Shear by a factor of +1.5", np.dot(cat, np.array([[1, 1.5], [0, 1]]))),

    ("Vertical Shear by a factor of -1.5", np.dot(cat, np.array([[1, 0], [-1.5, 1]]))),

    ("Vertical Shear by a factor of +1.5", np.dot(cat, np.array([[1, 0], [1.5, 1]]))),

    ("Projection onto the x-axis", np.dot(cat, np.array([[1, 0], [0, 0]]))),

    ("Projection onto the y-axis", np.dot(cat, np.array([[0, 0], [0, 1]]))),

    ("Rotation by an angle of π/3 counter-clockwise", np.dot(cat, np.array([[np.cos(-np.pi/3), -np.sin(-np.pi/3)], [np.sin(-np.pi/3), np.cos(-np.pi/3)]]))),

    ("Rotation by an angle of π/3 clockwise", np.dot(cat, np.array([[np.cos(np.pi/3), -np.sin(np.pi/3)], [np.sin(np.pi/3), np.cos(np.pi/3)]])))

]

for i, (title, transformation) in enumerate(transformations):
    ax = axs[i // 4, i % 4]
    ax.plot(transformation[:, 0], transformation[:, 1], 'b')
    ax.set_xlim(-2,2)
    ax.set_ylim(-2,2)
    ax.set_title(title,fontsize=8)

plt.tight_layout()
plt.show()
```

## Question 3

```python
from decimal import Decimal

def perform_operations(x, y):
    # Using float data type
    float_x = float(str(x))
    float_y = float(str(y))
    float_result_addition = float_x + float_y
    float_result_multiplication = float_x * float_y
    float_result_division = float_x / float_y

    # Using Decimal data type
    decimal_x = Decimal(str(x))
    decimal_y = Decimal(str(y))
    decimal_result_addition = decimal_x + decimal_y
    decimal_result_multiplication = decimal_x * decimal_y
    decimal_result_division = decimal_x / decimal_y

    # Display results and differences in precision
    print("Using float:")
    print(f"Addition: {float_result_addition}")
    print(f"Multiplication: {float_result_multiplication}")
    print(f"Division: {float_result_division}")

    print("\nUsing Decimal:")
    print(f"Addition: {decimal_result_addition}")
    print(f"Multiplication: {decimal_result_multiplication}")
    print(f"Division: {decimal_result_division}")

# User input
x = input("Enter the first number (x): ")
y = input("Enter the second number (y): ")
print()

perform_operations(x, y)

# (b)
# Use float for general-purpose programming where high precision is not critical.
#     It is suitable for most everyday calculations and offers good performance.
# Use Decimal for financial applications, scientific calculations, and situations
#     where precise decimal representation is required. It is essential for maintaining
#     accuracy in financial calculations, like currency conversions, where rounding errors
#     can lead to significant discrepancies.

(c)
# This result occurs due to the limitations of the float data type,
# which uses binary floating-point representation. In binary, the decimal value 0.1
# cannot be precisely represented, leading to small rounding errors in each addition.
# These accumulated rounding errors make the left-hand side of the equality (0.1 + 0.1 + 0.1)
# slightly different from the right-hand side (0.3), resulting in False.
# To achieve precise decimal arithmetic, you should use the Decimal data type.
```

## Question 4

```python
import math
import matplotlib.pyplot as plt

# Function to calculate π approximation using the Leibniz formula
def calculate_pi_approximation(N):
    pi_approximation = 0
    for k in range(N):
        # Alternating series: Add or subtract based on the index (k)
        pi_approximation += ((-1) ** k) / (2 * k + 1)
    return 4 * pi_approximation

# Function to calculate absolute error between computed approximation and known π
def calculate_absolute_error(approximation, known_value):
    return abs(approximation - known_value)

# Known value of π from the math module
known_pi = math.pi

# List to store N values for plotting
N_values = list(range(1, 1001))  # You can adjust the range as needed

# Calculate π approximations and absolute errors for different N values
pi_approximations = [calculate_pi_approximation(N) for N in N_values]
absolute_errors = [calculate_absolute_error(approx, known_pi) for approx in pi_approximations]

# Create a plot to visualize the evolution of absolute error with N
plt.figure(figsize=(10, 6))
plt.plot(N_values, absolute_errors)
plt.xlabel("Number of Terms (N)")
plt.ylabel("Absolute Error")
plt.title("Absolute Error vs. Number of Terms in Leibniz Formula")
plt.grid(True)
plt.show()
```
