## Manupulating Data

```python
car_sales['Price'] = car_sales['Price'].str.replace('[\$\,\.]', '', regex=True).astype(int)
car_sales
```

```python
car_sales.Price = car_sales['Price']/100
car_sales.Price.plot()
```

```python
car_sales.Make = car_sales.Make.str.lower()
```

```python
car_sales
```

```python
car_sales_missing = pd.read_csv("car-sales-missing-data.csv")
car_sales_missing
```

```python
car_sales_missing['Odometer'].fillna(car_sales_missing.Odometer.mean(), 
                                     inplace=True)
car_sales_missing
```

```python
car_sales_missing.dropna(inplace=True)
```

```python
car_sales_missing
```

```python
# Adding new column to dataframe
seats_col = pd.Series([5, 5, 5, 5])

car_sales['Seats'] = seats_col
car_sales.Seats.fillna(5, inplace=True)
car_sales
```

```python
# Column from python list
fuel_eco = [7.5, 9.2, 5.0, 9.6, 8.7, 5.0, 9.6, 8.7, 5.6, 7.3] # length must be same
car_sales['Fuel Economy'] = fuel_eco
car_sales['Number of Wheels'] = 4
car_sales['Passed road safety'] = True
car_sales
```

```python
car_sales.drop("Number of Wheels", axis=1, inplace=True)
car_sales
```

```python
# Sample
car_sales.sample(frac=1)
```

```python
car_sales.reset_index(drop=True)
```

```python
## Apply a function to column
car_sales["Odometer (mile)"] = car_sales['Odometer (KM)'].apply(lambda x: x/1.6)
```
